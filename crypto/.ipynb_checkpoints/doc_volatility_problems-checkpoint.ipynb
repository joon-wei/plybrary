#The problem with volatitlity.
When simulating a trade with ohlc data, which value should be used to trigger TP/SL?

Using 'High' and 'Low' values would be more accurate in simulating the volatility experienced in between candlesticks. However, in simulation, it lacks precision in which the order of which prices moved within the candle. 
i.e. If both TP and SL are within range of the candle, don't know which one hit first.
and volatile candles will instantly close a trade when it might not act that way in live markets.

Because both high and low wicks are touching TP and SL (even with a 20% and 25% range respectively), the trade was instantly stopped.

##Using High and Low:
Pros:
Simulates realistic intra-candle possibility of TP/SL being hit.
Conservative in assuming the trade is closed at the first touch.

Cons:
Too aggressive — ends trades quickly due to volatility.
Assumes instant fills at intrabar levels, which is not always achievable in real-world trading.
Lacks temporal sequencing within the candle.

##Alternative: Use the Close price
Pros:
Simpler and more stable simulations.
Reduces noise from intrabar volatility.

Cons:
Not realistic for intraday exits — in real trading, you don’t wait until close to act.
You may miss valid trades that would have hit TP or SL intrabar.

##Best Practice / More Accurate Simulation
Use more granular data if available:
1-minute or tick data lets you simulate actual order of price movement.
This solves the problem of ambiguity (which came first, TP or SL?).

Use sequential logic:
Always assume SL hits first before TP, for more conservative results.

Use Close as signal trigger only. Less sensitive to noise, but does not capture volatility which is important in volatile markets.

Consider adding a holding period.
